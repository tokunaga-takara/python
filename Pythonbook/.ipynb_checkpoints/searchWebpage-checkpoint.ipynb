```python
#スクレイピング
#requestモジュールのインポート
import requests
#reモジュールのインポート
import re
#Web上のファイルを取得
r = requests.get('https://www.python.org/downloads/')
#内容を表示
#print(r.text)

release = []
#htmlのタグで検索
for li in re.findall(r'<li>.+?</li>', r.text.replace('\n', '')): # マッチしたすべての部分を取得
    if x := re.search(r'<span class="release-number"><a href=".+?">(.+?)</a></span>', li): # 文字列中の位置にかかわらずにマッチ
        if y := re.search(r'<span class="release-date">(.+?)</span>', li): # 文字列中の先頭でのみマッチ
            release.append((x.group(1), y.group(1)))

release.sort()
for name, date in release:
    print(f'{name:15}{date}')
```

```python
#スクレイピング
#requestモジュールのインポート
import requests
#reモジュールのインポート
import re
#Web上のファイルを取得
r = requests.get('https://www.python.org/downloads/')
#内容を表示
#print(r.text)

search = []
#Pythonを含む文を検索
for s in re.findall(r'.+Python.+', r.text):
    search.append(s)

release.sort()
for s in search:
    print(s)
```
